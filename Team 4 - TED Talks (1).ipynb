```python
# Start your codes here!
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
import ast

```

```python
data = pd.read_csv("ted.csv")
data.head(5)
```

```python
data.dropna()
```

```python
data['film_date'] = data['film_date'].apply(lambda x: datetime.datetime.fromtimestamp( int(x)).strftime('%d-%m-%Y'))
data['published_date'] = data['published_date'].apply(lambda x: datetime.datetime.fromtimestamp( int(x)).strftime('%d-%m-%Y'))
```

```python
data = data[['title','description','name','main_speaker','speaker_occupation','event','film_date','published_date','languages','duration','related_talks','ratings','tags','views','comments','url']]
```

```python
data['funny'] = data['ratings'].apply(lambda x: x[0]['count'])
data['jawdrop'] = data['ratings'].apply(lambda x: x[-3]['count'])
data['beautiful'] = data['ratings'].apply(lambda x: x[3]['count'])
data['confusing'] = data['ratings'].apply(lambda x: x[2]['count'])
data.head()
```
